```python
# Renaming some of the seaons

def rename_seasons(value):
    
    # Looking at the columns, the season doesn't specify information about sowing/harvesting in season column.
    # Hence assuming following information.
    # Crops that are sown in monsoon(June - Sept) and harvested in autumn are combined under Monsoon
    if (value == "Kharif") | (value == "Autumn"):
        return "Monsoon"
    # Crops that are sown in Winter(Oct-Dec) and harvested in spring are combined under Winter
    elif value == "Rabi":
        return "Winter"
    else:
        return value
```

```python
a_data.season = a_data.season.apply(rename_seasons)
```

# Treating Nans in crop column

```python
# Decision: Based on above filter result, replacing the Nan crop values with Gram except for year 2011. 
# For year 2011 the major crop sown were groundnut.

df = a_data[(a_data.crop.isnull()) & (a_data.crop_year == 2011)]
df.crop = "Groundnut"

# Based on above result for district 'Tiruppur', yield is 6.15. Hence 2 * 6.15
df.production = 12.3
```

```python
a_data.crop = a_data.crop.fillna("Gram")
```

```python
# Correcting state names

rainfall_data.state = np.where((rainfall_data.state == "Orissa"), "Odisha", rainfall_data.state)
rainfall_data.state = np.where((rainfall_data.state == "Himachal"), "Himachal pradesh", rainfall_data.state)

rainfall_data.state = np.where((rainfall_data.state == "Pondicherry"), "Puducherry", rainfall_data.state)
rainfall_data.state = np.where((rainfall_data.state == "Daman and dui"), "Daman and diu", rainfall_data.state)

rainfall_data.state = np.where((rainfall_data.state == "Chatisgarh"), "Chhattisgarh ", rainfall_data.state)
rainfall_data.state = np.where((rainfall_data.state == "Uttaranchal"), "Uttarakhand", rainfall_data.state)

rainfall_data.state = np.where((rainfall_data.state == "Dadar nagar haveli"), "Dadra and nagar haveli", rainfall_data.state)

```

```python
rainfall_data["monsoon"] = rainfall_data.loc[:,["jun", "jul", "aug", "sep"]].mean(axis = 1)

rainfall_data["autumn"] = rainfall_data.loc[:,["oct"]]

rainfall_data["winter"] = rainfall_data.loc[:,["nov", "dec", "jan"]].mean(axis = 1)

rainfall_data["spring"] = rainfall_data.loc[:,["feb"]]

rainfall_data["summer"] = rainfall_data.loc[:,["mar", "apr", "may"]].mean(axis = 1)

rainfall_data["annual_mean"] = rainfall_data.loc[:,["annual"]]/12

rainfall_data
```

```python
# Exporting rainfall dataframe to .csv

rainfall_data.to_csv("data/clean/Rainfall_clean.csv", index = False)
```
